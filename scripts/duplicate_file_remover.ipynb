```python
import os

def find_and_delete_duplicate_files(directory):
    files = sorted(os.listdir(directory))

    i = 0
    while i < len(files) - 1:
        file1 = files[i]
        file2 = files[i + 1]

        # # Check if the file names have at least two "_" characters
        if file1.count("_") >= 2 and file2.count("_") >= 2:
        #     # Get the substring before the second "_" in each file name
            name1, _, suffix1 = file1.partition("_")[2].partition("_")
            name2, _, suffix2 = file2.partition("_")[2].partition("_")

        #     # Compare if the file names are the same before the second "_"
            if name1 == name2:
        #         # Delete the latter file
                file_to_delete = os.path.join(directory, file2)
                os.remove(file_to_delete)
                print(f"Deleted file: {file_to_delete}")

        #         # Since we deleted a file, we don't increment i to check the next file
        #         # Instead, we'll re-check the current file with the one that comes after

        i += 1
```

```python
target_directory = "/home/semantic-server/soganghci/api-server/src/documents/wikipedia"
find_and_delete_duplicate_files(target_directory)
```
